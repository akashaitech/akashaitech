```python
!pip install rarfile
from google.colab import files
import os
import rarfile
import pandas as pd
import cv2
import numpy as np
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import RandomOverSampler # For oversampling
from collections import Counter

# Step 1: Upload Files
print(" Please upload 'heart.csv' and 'ECG Dataset.rar'")
uploaded = files.upload()

# Step 2: Extract ECG Dataset
rar_file = next((f for f in uploaded if f.endswith('.rar')), None)
extract_folder = 'ECG_Dataset'
os.makedirs(extract_folder, exist_ok=True)

if rar_file:
    with rarfile.RarFile(rar_file) as rf:
        rf.extractall(extract_folder)
    print(" ECG Dataset extracted!")
else:
    print(" No .rar file found.")
    exit()

# Step 3: Load and Preprocess Tabular Data (heart.csv)
csv_file = next((f for f in uploaded if f.endswith('.csv')), None)
if csv_file:
    heart_df = pd.read_csv(csv_file)
    print(" heart.csv loaded successfully! Shape:", heart_df.shape)
    # Select numerical features, INCLUDING 'target' for separate training
    numerical_features = heart_df.select_dtypes(include=np.number).columns.tolist()
    if 'target' in numerical_features:
        tabular_features = heart_df[numerical_features].dropna(axis=1)
        print(" Selected Tabular Features (including target) Shape:", tabular_features.shape)
        X_tab = tabular_features.drop('target', axis=1)
        y_tab = tabular_features['target']
        X_tab_scaled = StandardScaler().fit_transform(X_tab)
        X_train_tab, X_test_tab, y_train_tab, y_test_tab = train_test_split(
            X_tab_scaled, y_tab, test_size=0.2, random_state=42, stratify=y_tab
        )
        print(" Tabular Data split into training and testing sets.")
    else:
        print(" 'target' column not found in heart.csv. Cannot train tabular model separately.")
        X_train_tab, X_test_tab, y_train_tab, y_test_tab = None, None, None, None
else:
    print(" No .csv file found. Proceeding with only image data.")
    X_train_tab, X_test_tab, y_train_tab, y_test_tab = None, None, None, None

# Step 4: Load and Preprocess ECG Images
image_data = []
base_path = os.path.join(extract_folder, 'ECG Dataset')

for root, _, files in os.walk(base_path):
    label = os.path.basename(root)
    if label in ['Abnormal heartbeat', 'History of MI', 'Normal Person']:
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                image_data.append({'filepath': os.path.join(root, file), 'label': label})

ecg_df = pd.DataFrame(image_data)
print(f" Total ECG images found: {len(ecg_df)}")

def preprocess_image(filepath, size=(128, 128)):
    try:
        img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, size)
        return img / 255.0
    except:
        return None

ecg_df['image_array'] = [preprocess_image(path) for path in tqdm(ecg_df['filepath'])]
ecg_df = ecg_df[ecg_df['image_array'].notnull()]

X_images_raw = np.stack(ecg_df['image_array'].values).reshape(-1, 128, 128, 1)
le_img = LabelEncoder()
y_labels_raw_img = le_img.fit_transform(ecg_df['label'])

print(" Raw Images shape:", X_images_raw.shape)
print(" Raw Image Labels shape:", y_labels_raw_img.shape)
print(" Image Label classes:", le_img.classes_)
print(" Image Class distribution before oversampling:", Counter(y_labels_raw_img))

# Step 5: Balance Image Data using Oversampling
ros_img = RandomOverSampler(random_state=42)
X_images_resampled, y_labels_resampled_img = ros_img.fit_resample(X_images_raw.reshape(X_images_raw.shape[0], -1), y_labels_raw_img)
X_images_resampled = X_images_resampled.reshape(-1, 128, 128, 1)

print(" Resampled Images shape:", X_images_resampled.shape)
print(" Resampled Image Labels shape:", y_labels_resampled_img.shape)
print(" Image Class distribution after oversampling:", Counter(y_labels_resampled_img))

X_train_img, X_test_img, y_train_img, y_test_img = train_test_split(
    X_images_resampled, y_labels_resampled_img, test_size=0.2, random_state=42, stratify=y_labels_resampled_img
)
print(" Image Data split into training and testing sets.")

# Step 6: Train Model on Tabular Data (if available)
if X_train_tab is not None:
    tabular_model = Sequential([
        Dense(128, activation='relu', input_shape=(X_train_tab.shape[1],)),
        Dropout(0.5),
        Dense(len(np.unique(y_train_tab)), activation='softmax') # Use number of unique tabular labels
    ])
    tabular_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    history_tab = tabular_model.fit(X_train_tab, y_train_tab, epochs=20, batch_size=32, validation_data=(X_test_tab, y_test_tab))
    loss_tab, accuracy_tab = tabular_model.evaluate(X_test_tab, y_test_tab)
    print(f"\n Tabular Model Test Accuracy: {accuracy_tab * 100:.2f}%")

    y_pred_tab = np.argmax(tabular_model.predict(X_test_tab), axis=1)
    print("\n Classification Report (Tabular Model):")
    print(classification_report(y_test_tab, y_pred_tab))

    cm_tab = confusion_matrix(y_test_tab, y_pred_tab)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_tab, annot=True, fmt='d', cmap='Greens',
                xticklabels=np.unique(y_tab), yticklabels=np.unique(y_tab))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix (Tabular Model)')
    plt.show()
else:
    print("\n Tabular model training skipped.")

# Step 7: Train Model on Image Data
image_input = Input(shape=(128, 128, 1), name='image_input')
conv1_img = Conv2D(32, (3, 3), activation='relu')(image_input)
pool1_img = MaxPooling2D(2, 2)(conv1_img)
conv2_img = Conv2D(64, (3, 3), activation='relu')(pool1_img)
pool2_img = MaxPooling2D(2, 2)(conv2_img)
flatten_img = Flatten()(pool2_img)
dense1_img = Dense(128, activation='relu')(flatten_img)
dropout1_img = Dropout(0.5)(dense1_img)
output_img = Dense(len(le_img.classes_), activation='softmax')(dropout1_img)

image_model = Model(inputs=image_input, outputs=output_img)
image_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_img = image_model.fit(X_train_img, y_train_img, epochs=20, batch_size=32, validation_data=(X_test_img, y_test_img))
loss_img, accuracy_img = image_model.evaluate(X_test_img, y_test_img)
print(f"\n Image Model Test Accuracy: {accuracy_img * 100:.2f}%")

y_pred_img = np.argmax(image_model.predict(X_test_img), axis=1)
print("\n Classification Report (Image Model):")
print(classification_report(y_test_img, y_pred_img, target_names=le_img.classes_))

cm_img = confusion_matrix(y_test_img, y_pred_img)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_img, annot=True, fmt='d', cmap='Reds',
            xticklabels=le_img.classes_, yticklabels=le_img.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix (Image Model)')
plt.show()
```

```python
import numpy as np

# --- Class labels ---
tabular_classes = ["Heart Disease", "Normal"]
image_classes = ["Abnormal heartbeat", "History of MI", "Normal Person"]

# --- Get user input for tabular probabilities ---
print(" Enter tabular model probabilities:")
tabular_probs = []
for cls in tabular_classes:
    prob = float(input(f"Probability for '{cls}': "))
    tabular_probs.append(prob)
tabular_probs = np.array(tabular_probs)

# --- Get user input for image model probabilities ---
print("\n Enter image model probabilities:")
image_probs = []
for cls in image_classes:
    prob = float(input(f"Probability for '{cls}': "))
    image_probs.append(prob)
image_probs = np.array(image_probs)

# --- Final decision logic ---
def final_prediction(tabular_probs, image_probs, tabular_classes, image_classes):
    tabular_class = tabular_classes[np.argmax(tabular_probs)]
    image_class = image_classes[np.argmax(image_probs)]

    print(f"\n Tabular Predicted Class: {tabular_class} (Probs: {tabular_probs})")
    print(f" Image Predicted Class: {image_class} (Probs: {image_probs})")

    if tabular_class == "Heart Disease" and image_class == "Abnormal heartbeat":
        return " High probability of heart attack"
    elif tabular_class == "Normal" and image_class == "Normal Person":
        return " Person is healthy"
    elif tabular_class == "Normal" and image_class == "History of MI":
        return " Person has heart problem but not at risk currently"
    elif tabular_class == "Heart Disease" and image_class == "History of MI":
        return " Person has heart problem and at risk currently"
    else:
        return " Undecided"

# --- Run prediction ---
final_class = final_prediction(tabular_probs, image_probs, tabular_classes, image_classes)
print(f"\n Final Predicted Class: {final_class}")
```
